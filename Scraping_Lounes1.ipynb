```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd
```

```python
# Garder la page ouverte
chrome_options = Options() 
chrome_options.add_experimental_option("detach", True)

# On recupere le controle du driver
DRIVER_PATH = "./chromedriver"
driver = webdriver.Chrome(DRIVER_PATH, options=chrome_options)
driver.implicitly_wait(10)
```

```python
# On va chehcher notre site web
page_url = "https://www.alvergnas.com/trouver-un-vehicule/#!/vehicle?pg="
driver.get(page_url)
time.sleep(5)
```

```python
#recuperation du nombre d'element

nbr = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div/div/p/app-root/div/app-list/div[1]/div[2]/mat-paginator/div/div[2]/div').text


nbr_elems = ""
for i in range(len(nbr)-1, 0, -1):
    if nbr[i]== " ":
        break
    nbr_elems = nbr[i] + nbr_elems

print(nbr_elems)

#recuperation du nombre de page
nbr_pages = int(int(nbr_elems) / 12)
print(nbr_pages)

#recuperation du nombre d'element sur la dernière page
nbr_elem_last_page = int(nbr_elems)-(12*nbr_pages)
print(nbr_elem_last_page)

driver.close()
driver.quit()
```

```python
#Creation des tableaux qui ou seront stockées les données de chaque voiture
name=[]#
price=[]#
loca = []#
cate = []#
fuel = []#
power = []#
doors = []#
couleur_interieur = []#
space = []#
premiere_main = []#
annee = []#
kilometrage = []#
boite_de_vitesse = []#
puissance_fiscale = []#
couleur_exterieur = []#
inte = []#
date_de_distribution = []#
tva_recuperable = []
```

```python
elem = 12

for a in range(1, nbr_pages+1):
#for a in range(1,3):

        # On reouvre Chrome a chaque nouvelle page

        # Garder la page ouverte
        chrome_options = Options() 
        chrome_options.add_experimental_option("detach", True)

        # On recupere le controle du driver
        DRIVER_PATH = "./chromedriver"
        driver = webdriver.Chrome(DRIVER_PATH, options=chrome_options)
        driver.implicitly_wait(10)


        if a == nbr_pages:
                elem = nbr_elem_last_page

        page_url = "https://www.alvergnas.com/trouver-un-vehicule/#!/vehicle?pg=" + str(a)
        driver.get(page_url)
        time.sleep(5)
        
        #print(a)
        #print(elem)

        for i in range(elem):

                #print(i)
                try :
                        driver.find_element(By.XPATH,f"//*[@id='et-boc']/div/div/div/div/div/div/p/app-root/div/app-list/div[2]/div/div[{i+1}]/mat-card/mat-card-actions/button").click()
                        time.sleep(5)
                        
                        try:
                                nom =  driver.find_element(By.XPATH, "//*[@id='print-section']/div[1]/div[2]/div[1]/h1").text
                        except:
                                nom = None
                        try:  
                                prix = driver.find_element(By.XPATH, "//*[@id='print-section']/div[1]/div[3]/div[2]/div[2]/div/button").text
                        except:
                                prix = None
                        try:
                                catégorie = driver.find_element(By.XPATH, "//*[@id='print-section']/div[1]/div[5]/span[1]/div[2]/span").text
                        except:
                                catégorie = None
                        try :
                                bdv =  driver.find_element(By.XPATH,'//*[@id="print-section"]/div[1]/div[5]/span[4]/span/div[2]/span').text
                        except:
                                bdv = None
                        try:
                                portes = driver.find_element(By.XPATH,'//*[@id="print-section"]/div[1]/div[5]/span[7]/span/div[2]/span').text
                        except:
                                portes = None
                        try:
                                intérieur =  driver.find_element(By.XPATH,'//*[@id="print-section"]/div[1]/div[5]/span[10]/span/div[2]/span').text
                        except:
                                intérieur = None
                        try:
                                pm =  driver.find_element(By.XPATH,'//*[@id="print-section"]/div[1]/div[5]/span[13]/span/div[2]/span').text
                        except:
                                pm = None
                        try:
                                km = driver.find_element(By.XPATH,'//*[@id="print-section"]/div[1]/div[5]/span[2]/span/div[2]/span').text
                        except:
                                km = None
                        try:
                                puissance = driver.find_element(By.XPATH,'//*[@id="print-section"]/div[1]/div[5]/span[5]/span/div[2]/span').text
                        except:
                                puissance = None
                        try:
                                couleur = driver.find_element(By.XPATH,'//*[@id="print-section"]/div[1]/div[5]/span[8]/span/div[2]/span').text
                        except:
                                couleur = None
                        try:
                                places = driver.find_element(By.XPATH,'//*[@id="print-section"]/div[1]/div[5]/span[11]/span/div[2]/span').text
                        except:
                                places = None
                        try:
                                tva = driver.find_element(By.XPATH,'//*[@id="print-section"]/div[1]/div[5]/span[14]/span/div[2]/span').text
                        except:
                                tva = None
                        try:
                                carburant = driver.find_element(By.XPATH,'//*[@id="print-section"]/div[1]/div[5]/span[3]/span/div[2]/span').text
                        except:
                                carburant = None
                        try:
                                puissancef = driver.find_element(By.XPATH,'//*[@id="print-section"]/div[1]/div[5]/span[6]/span/div[2]/span').text
                        except:
                                puissancef = None
                        try:
                                datedistrib = driver.find_element(By.XPATH,'//*[@id="print-section"]/div[1]/div[5]/span[12]/span/div[2]/span').text
                        except:
                                datedistrib = None
                        try:
                                date = driver.find_element(By.XPATH,'//*[@id="print-section"]/div[1]/div[5]/span[15]/span/div[2]/span').text
                        except:
                                date = None
                        try:
                                location = driver.find_element(By.XPATH,'//*[@id="print-section"]/div[1]/div[3]/div[2]/div[3]/div[1]/h2/span').text
                        except:
                                location = None
                        try:
                                couleurI = driver.find_element(By.XPATH,'//*[@id="print-section"]/div[1]/div[5]/span[9]/span/div[2]/span').text
                        except:
                                couleurI = None


                        price.append(prix)
                        #print(price)
                        name.append(nom)
                        #print(name)
                        cate.append(catégorie)
                        #print(cate)
                        boite_de_vitesse.append(bdv)
                        #print(boite_de_vitesse)
                        doors.append(portes)
                        #print(doors)
                        inte.append(intérieur)
                        #print(inte)
                        premiere_main.append(pm)
                        #print(premiere_main)
                        power.append(puissance)
                        #print(power)
                        couleur_exterieur.append(couleur)
                        #print(couleur_exterieur)
                        space.append(places)
                        #print(space)
                        tva_recuperable.append(tva)
                        #print(tva_recuperable)
                        fuel.append(carburant)
                        #print(fuel)
                        puissance_fiscale.append(puissancef)
                        #print(puissance_fiscale)
                        date_de_distribution.append(datedistrib)
                        #print(date_de_distribution)
                        annee.append(date)
                        #print(annee)
                        loca.append(location)
                        #print(loca)
                        kilometrage.append(km)
                        #print(kilometrage)
                        couleur_interieur.append(couleurI)
                        #print(couleur_interieur)

                        driver.back()
                        time.sleep(5)
                except:
                        pass
        

        # On ferme Chrome a chaque fin de page
        driver.close()
        driver.quit()

```
